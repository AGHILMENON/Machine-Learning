```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid=SentimentIntensityAnalyzer()
```

```python
emo.head()
```

```python
emo.loc[0]['Text']
```

```python
sid.polarity_scores(emo.loc[0]['Text'])
```

```python
emo['Score']=emo['Text'].apply(lambda scores:sid.polarity_scores(scores))
```

```python
emo
```

```python
emo['compound']  = emo['Score'].apply(lambda scores: scores['compound'])
emo.head()
```

```python
emo['Score'][0]['compound']
```

```python
emo['comp_score'] = emo['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
emo
```
